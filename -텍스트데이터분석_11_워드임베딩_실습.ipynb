# 영화 리뷰 워드 임베딩 (Word2Vec, FastText)
- gensim 라이브러리 사용 : pip install gensim
    - Word2Vec : models.Word2Vec
    - FastText : models.FastText

## 1. 데이터 준비
* 토큰화가 잘 되어 있는 filtered 데이터 사용

```python
import pandas as pd
data_filename = './data/Korean_movie_reviews_2016_filtered.csv'
data_df = pd.read_csv(data_filename)
data_df.head()
```

```python
data_df.info()
```

```python
#   결측치 제거
data_df.dropna(inplace = True)
```

```python
#   결측치 제거 (rate에 결측치)
#   결측치를 제거하지 않으면 review 데이터 추출 시 float으로 타입 변환됨
data_df.info()
```

```python
# review만 모아서 review별 토큰 리스트로 변환
review_list = list(data_df.review)
review_list[:5]
```

```python
type(review_list[0])
```

```python
#   결측치 처리 시 AttributeError 예외 발생
token_list = [review.split() for review in review_list ]
print(token_list[:2])
```

## 1. Word2Vec 활용 영화 리뷰 워드 임베딩
* https://radimrehurek.com/gensim/models/word2vec.html

### Skipgram, negative=10 인 경우

```python
# Word2Vec 모델 생성 및 학습 : window=3, min_count=3
from gensim.models import Word2Vec
model_sg_n10 = Word2Vec(token_list, min_count = 3, vector_size = 100, sg = 1, negative = 10, window = 3)
#   fit 개념 없음, 파라미터에 맞춰서 단어 만들어 줌
```

```python
# 단어의 임베딩 벡터 확인
model_sg_n10.wv['이정재']
```

```python
# 단어의 임베딩 벡터 차원 확인
len(model_sg_n10.wv['이정재'])
```

```python
# 두 단어 간 유사도 확인
wv = model_sg_n10.wv
wv.similarity('이정재', '정우성')
```

```python
# 특정 단어와 유사한 단어 추출
wv.most_similar('이정재', topn = 20)
```

```python
print([word for word, _ in wv.most_similar('재밌', topn = 50)])
```

### Skipgram, negative=5 인 경우

```python
# 모델 생성
model_sg_n5 = Word2Vec(token_list, min_count = 3, vector_size = 100, sg = 1, negative = 5, window = 3)
```

```python
# 특어 단어와 유사한 단어 추출 : 이정재
wv = model_sg_n5.wv
wv.most_similar('이정재', topn = 20)
```

```python
# 특어 단어와 유사한 단어 추출 : 재밌
print([word for word, _ in wv.most_similar('재밌', topn = 50)])
```

### CBOW, negative=10 인 경우

```python
model_cbow_n10 = Word2Vec(token_list, min_count = 3, vector_size = 100, sg = 0, negative = 10, window = 3)
```

```python
wv = model_cbow_n10.wv
wv.most_similar('이정재', topn = 20)
```

```python
print([word for word, _ in wv.most_similar('재밌', topn = 50)])
```

### CBOW, negative=5 인 경우

```python
model_cbow_n5 = Word2Vec(token_list, min_count = 3, vector_size = 100, sg = 0, negative = 5, window = 3)
```

```python
wv = model_cbow_n5.wv
wv.most_similar('이정재', topn = 20)
```

```python
print([word for word, _ in wv.most_similar('재밌', topn = 50)])
```

### OOV(Out of Vocabulary) 문제

```python
# corpus에 없는 단어 확인 : 우주평화 
'우주평화' in model_sg_n10.wv.key_to_index
```

```python
# corpus에 없는 단어의 임베딩 벡터 확인 
model_sg_n10.wv['우주평화']
```

## 2. FastText 활용 영화 리뷰 워드 임베딩
* https://radimrehurek.com/gensim/models/fasttext.html

```python
# FastText 모델 생성 및 학습
# window=3, min_count=3, min_n=2, max_n=2
from gensim.models import FastText
model = FastText(token_list, min_count = 3, vector_size = 100, sg = 1, negative = 10, window = 3, min_n = 2, max_n = 2)
wv = model.wv
```

```python
# 특정 단어와 유사한 단어 추출 : 이정재
wv.most_similar('이정재', topn = 20)
```

```python
# corpus에 없는 단어 확인 : 우주평화 
'우주평화' in model_sg_n10.wv.key_to_index
```

```python
# corpus에 없는 단어의 임베딩 벡터 확인 
wv['우주평화']
```

```python
# corpus에 없는 단어와 유사한 단어추출 
wv.most_similar('우주평화', topn = 20)
```
