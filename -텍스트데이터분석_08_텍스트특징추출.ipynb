# 텍스트 특징 추출 (TF-IDF)

### 1. 샘플 코퍼스로 연습

```python
sample_corpus = [
    '자연어처리 강의를 시작하겠습니다.',
    '자연어처리는 재미있습니다.',
    '밥을 먹고 강의를 듣고 있습니다.',
    '이번 자연어처리 강의는 한국어 자연어처리입니다.'
]
```

```python
import sklearn
```

```python
from konlpy.tag import Okt

def my_tokenizer(text):
    return Okt().nouns(text)
```

```python
# tfidVectorizer 객체 생성
#from sklearn.feature_extraction.text import TfidfVectorizer
#vectorizer = TfidfVectorizer(tokenizer = my_tokenizer)

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer = my_tokenizer)

# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(sample_corpus)
print(vectorizer.get_feature_names_out())
# 특징 벡터 추출(빈도 수 및 문서必)
sample_dtm = vectorizer.transform(sample_corpus)
type(sample_dtm)#   sparse array 타입
sample_dtm.toarray()#   일반 list형태로
```

### 2. 다음 영화 리뷰로 연습

```python
#   max() token 확인 후 분석 진행
import os
import pandas as pd
import sklearn
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter

tokenizer = Okt()

def my_tokenizer(text):
    return Okt().nouns(text)

file_path = r'D:\Lecture\TextMining26\data'
file_name = 'daum_movie_review.csv'

path_target = os.path.join(file_path, file_name)

movie_review = pd.read_csv(path_target, dtype='str')
```

```python
print(movie_review.head())
```

```python
#   한글만 남기고
movie_review['set_review'] = movie_review["review"].str.replace("[^ㅏ-ㅣㄱ-ㅎ가-힣]+", " ", regex=True)
```

```python
result = []

for word in movie_review['set_review']:
    result.append(word)

movie_review['tokens'] = result
```

```python
all_token = [word for tokens in movie_review['tokens'] for word in tokens]
count_tokens = Counter(all_token)

print(len(count_tokens))
```

```python
stop_tokens = []

vectorizer = CountVectorizer(tokenizer = my_tokenizer,
                             stop_words = stop_tokens,
                             max_features = 1000,
                             min_df = 2,
                             max_df = 0.8
                             )
```

```python
# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(movie_review['set_review'])
print(vectorizer.get_feature_names_out())

sample_dtm = vectorizer.transform(movie_review['set_review'])
print(sample_dtm.toarray())
```
